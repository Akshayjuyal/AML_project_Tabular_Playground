```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
df = pd.read_csv('train_0_truncated_tail.csv')

df.pop('boost0_timer')
df.pop('boost1_timer')
df.pop('boost2_timer')
df.pop('boost3_timer')
df.pop('boost4_timer')
df.pop('boost5_timer')

df.pop('p0_boost')
df.pop('p1_boost')
df.pop('p2_boost')
df.pop('p3_boost')
df.pop('p4_boost')
df.pop('p5_boost')

df.pop('p0_pos_z')
df.pop('p1_pos_z')
df.pop('p2_pos_z')
df.pop('p3_pos_z')
df.pop('p4_pos_z')
df.pop('p5_pos_z')

df.pop('p0_vel_z')
df.pop('p1_vel_z')
df.pop('p2_vel_z')
df.pop('p3_vel_z')
df.pop('p4_vel_z')
df.pop('p5_vel_z')

df.pop('ball_pos_z')
df.pop('ball_vel_z')

df.head()

event_time = df["event_time"]
ball_x = df["ball_pos_x"]
ball_y = df["ball_pos_y"]

ball_vel_x = df["ball_vel_x"]
ball_vel_y = df["ball_vel_y"]

min_time = min(event_time)
max_time = max(event_time)


#Since Player Scoring Next in traning data is player 3, let's analyze their movement 

p3_x = df["p3_pos_x"]
p3_y = df["p3_pos_y"]

p3_vel_x = np.array(df["p3_vel_x"])
p3_vel_y = np.array(df["p3_vel_y"])

p3_vel_mag = np.zeros(len(p3_vel_x))
for i in range(len(p3_vel_x)):
  p3_vel_mag[i] = math.sqrt(p3_vel_x[i]**2+p3_vel_x[i]**2)


ball_vel_x = np.array(df["ball_vel_x"])
ball_vel_y = np.array(df["ball_vel_y"])

ball_vel_mag = np.zeros(len(ball_vel_x))
for i in range(len(ball_vel_x)):
  ball_vel_mag[i] = math.sqrt(ball_vel_x[i]**2+ball_vel_x[i]**2)




df.head()
```

```python
plt.title("Ball Velocity Magnitude from {} to {}".format(min_time, max_time))
plt.xlabel("Event Time")
plt.ylabel("Ball Velocity Magnitude")
plt.scatter(event_time, ball_vel_mag)
```

```python
plt.title("Ball x Position from {} to {}".format(min_time, max_time))
plt.xlabel("time")
plt.ylabel("Ball x Position")
plt.scatter(event_time, ball_x)
```

```python
plt.title("Ball y Position from {} to {}".format(min_time, max_time))
plt.xlabel("time")
plt.ylabel("Ball y Position")
plt.scatter(event_time, ball_y)
```

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.plot3D(event_time, ball_x, ball_y)
ax.set_title("Ball x,y Position from {} to {}".format(min_time, max_time))
plt.show()
```

```python
ball_vel_x = df["ball_vel_x"]
plt.xlabel("time")
plt.ylabel("Ball Velocity x Comp")
plt.title("Ball Velocity x Component from {} to {}".format(min_time, max_time))
plt.scatter(event_time, ball_vel_x)
```

```python
ball_vel_y = df["ball_vel_y"]
plt.xlabel("time")
plt.ylabel("Ball Velocity y Comp")
plt.title("Ball Velocity y Component from {} to {}".format(min_time, max_time))
plt.scatter(event_time, ball_vel_y)
```

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.plot3D(ball_vel_x, ball_vel_y, p3_vel_mag)
ax.set_xlabel('ball x vel')
ax.set_ylabel('ball y vel')
ax.set_zlabel('p3 vel mag')
ax.set_title("Player 3 Vel. Mag. WRT Ball Vector x,y from {} to {}".format(min_time, max_time))
plt.show()
```

```python
plt.title("Player 3 x position from {} to {}".format(min_time, max_time))
plt.xlabel("time")
plt.ylabel("player 3 y position")
plt.scatter(event_time, p3_x)
```

```python
plt.title("Player 3 y position from {} to {}".format(min_time, max_time))
plt.xlabel("time")
plt.ylabel("player 3 x position")
plt.scatter(event_time, p3_y)
```

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.plot3D(p3_x, p3_y,ball_x)
ax.set_xlabel('p3 x position')
ax.set_ylabel('p3 y position')
ax.set_zlabel('ball x')
ax.set_title("Ball X-Pos WRT Player 3 x,y Position from {} to {}".format(min_time, max_time))
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.plot3D(p3_x, p3_y,ball_y)
ax.set_xlabel('p3 x position')
ax.set_ylabel('p3 y position')
ax.set_zlabel('ball y')
ax.set_title("Ball Y-Pos WRT Player 3 x,y Position from {} to {}".format(min_time, max_time))
plt.show()
```

```python
plt.title("Player 4 Velocity Vector x Component from {} to {}".format(min_time, max_time))
plt.xlabel("time")
plt.ylabel("player 3 vel x comp")
plt.scatter(event_time, p3_vel_x)
```

```python
plt.title("Player 3 Velocity Vector y Component from {} to {}".format(min_time, max_time))
plt.xlabel("time")
plt.ylabel("player 3 vel y comp")
plt.scatter(event_time, p3_vel_y)
```

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.plot3D(p3_vel_x, p3_vel_y,ball_x)
ax.set_title("Ball X-Pos WRT Player 3 Velocity Vector x,y Components from {} to {}".format(min_time, max_time))
ax.set_xlabel('player 3 vel-x')
ax.set_ylabel('player 3 vel-y')
ax.set_zlabel('ball x')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.plot3D(p3_vel_x, p3_vel_y,ball_y)
ax.set_title("Ball Y-Pos WRT Player 3 Velocity Vector x,y Components from {} to {}".format(min_time, max_time))
ax.set_xlabel('velocity x component')
ax.set_ylabel('velocity y component')
ax.set_zlabel('ball y')
plt.show()
```

```python
plt.title("Ball Vel-Mag WRT P3 Vel-Mag from {} to {}".format(min_time, max_time))
plt.xlabel("P3 Velocity Mag.")
plt.ylabel("Ball Velocity Mag.")
plt.scatter(p3_vel_mag, ball_vel_mag, c=event_time, cmap='BuPu_r')
```

```python
# At what time are the ball and player's velocity magnitude large
# that means the ball is being hit at that time point

fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.plot3D(ball_vel_mag, p3_vel_mag,event_time)
ax.set_title("Time WRT P3 Vel-Mag, Ball Vel-Mag from {} to {}".format(min_time, max_time))
ax.set_xlabel('P3 Vel-Mag')
ax.set_ylabel('Ball Vel-Mag')
ax.set_zlabel('time')
plt.show()

plt.title("P3 Vel-Mag and Ball Vel-Mag from {} to {}".format(min_time, max_time))
plt.xlabel("P3 Vel-Mag")
plt.ylabel("Ball Vel-Mag")
plt.scatter(ball_vel_mag, p3_vel_mag, c=event_time, cmap='BuPu_r')
```
