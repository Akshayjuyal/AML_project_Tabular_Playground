```python
'''
CODE IMPORTS
'''
import numpy as np
import pandas as pd
import math
```

```python
'''
IMPORT TRAINING DATA FILE
AND PREVIEW THE UNMODIFIED DATA
'''
df = pd.read_csv('train_0_truncated_tail.csv')
print("df.iloc[0] before modification is {}".format(len(df.iloc[0])))

'''
REMOVE THE FOLLOWING FEATURES
(THEY ARE NOT IN THE TESTING DATA)
  1 game_num
  2 event_id
  3 event_time
  4 player_scoring_next
  5 team_scoring_next
  6 team_A_scoring_within_10sec
  7 team_B_scoring_within_10sec
'''


df.pop('game_num')
df.pop('event_id')
df.pop('event_time')
df.pop('player_scoring_next')
df.pop('team_scoring_next')
#df.pop('player_scoring_next') there is a problem when it is not commented out?
p_A_swts_gt = df.pop('team_A_scoring_within_10sec').astype(int)
p_B_swts_gt = df.pop('team_B_scoring_within_10sec').astype(int)

print("df.iloc[0] after modification is {}".format(len(df.iloc[0])))


'''
REPLACE ANY NANS WITH ZEROS
'''
df = df.fillna(0)
```

```python
#https://towardsdatascience.com/how-to-code-logistic-regression-from-scratch-with-numpy-d33c46d08b7f
'''
This article derives compact closed form formulas for the logistic regression
Use these to compute the parameters of the logistic regression
Compute Probability scores using logistic regression parameters
'''

'''
TAKE DATAFRAME AS NUMPY ARRAY
'''
X = np.array(df)


'''
MAKE LABELS LIST FROM team_A_scoring_within_10sec AND team_B_scoring_within_10sec...
  ...COLUMNS IN DATAFRAME
'''
p_B_swts_gt[ p_A_swts_gt == 1 ] = -1
Y = np.zeros(len(p_B_swts_gt)).astype(int)
for i in range(len(p_B_swts_gt)):
  if(p_A_swts_gt[i] != 0):
    Y[i] = 1
  if(p_B_swts_gt[i] != 0):
    Y[i] = 2
  if((p_A_swts_gt[i] == 0) and (p_B_swts_gt[i] == 0)):
    Y[i] = 3

'''
CREATE W VECTOR
'''
dim = len(X[0])
w = np.zeros(dim)
c = np.zeros(len(X))
mid = np.divide(1,Y)-np.ones(len(X))
for i in range(len(mid)):
  c[i] = math.log(abs(mid[i]))
res = c @ X
w = -np.linalg.inv(X.T @ X) @ X.T 
w = w.T @ res #w in wrong dim shape=1000

'''
COMPUTE SCORES USING LOGISTIC SIGMOID FUNCTION
'''
x_test = X
y_test = Y
y_pred = np.zeros((2,len(x_test)))
for i in range(len(x_test)):
  y_pred[0][i] = (math.e**(np.dot(w,x_test[i]))) / (1+math.e**(np.dot(w,x_test[i])))
  y_pred[1][i] = 1 /(math.e**-(np.dot(w,x_test[i])))


```
